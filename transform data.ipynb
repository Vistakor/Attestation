```python
import pandas as pd
```

```python
df = pd.read_csv('/content/KIA', delimiter=';', encoding='utf-8')
df.head()
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
df.head()
```

```python
df.dtypes
```

```python
df['модель'].value_counts()
```

```python
drops = ['XCEED', 'CARENS', 'CEED_GT', 'K8', 'K900', 'K7', 'MORNING', 'CLARUS', 'SHUMA', 'PRIDE']
for drop in drops:
  df.drop(df.index[df['модель'] == drop], inplace=True)
```

```python
df['страна сборки'].value_counts()
```

```python
df = df.drop('страна сборки', axis=1)
```

```python
df['модель'] = df['модель'].astype('category')
df['тип двигателя'] = df['тип двигателя'].astype('category')
df['коробка передач'] = df['коробка передач'].astype('category')
```

```python
df.head()
```

```python
df.dtypes
df.insert(1, 'модель кат', df['модель'].cat.codes)
df.insert(3, 'тип двигателя кат', df['тип двигателя'].cat.codes)
df.insert(5, 'коробка передач кат', df['коробка передач'].cat.codes)
df.head()
```

```python
df['без дтп'] = df['без дтп'].replace({ True : 1 , False : 0 })
df.head()
```

```python
df_model = df.drop(['модель', 'тип двигателя', 'коробка передач'], axis=1)
df_model.head()
```

```python
df_model.to_csv('df_model')
```
